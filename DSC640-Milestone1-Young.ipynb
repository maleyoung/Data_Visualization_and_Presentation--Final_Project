```python
import pandas as pd
```

```python
#load the Excel file into a DataFrame
df = pd.read_excel('nationaldatabaseofchildcareprices.xlsx')

#display the DataFrame
df.head()
```

```python
df.shape
```

```python
#filter the DataFrame to keep only rows where State_Name is Florida
florida_df = df[df['State_Name'] == 'Florida']

#display the filtered DataFrame
florida_df.head(50)
```

```python
florida_df.shape
```

```python
#show all columns
pd.set_option('display.max_columns', None) 
#show all rows
pd.set_option('display.max_rows', None) 
#prevent wrapping
pd.set_option('display.expand_frame_repr', False)  
#creating a list of the column names
florida_df.columns.tolist()  
#I will use this list of column names to filter only the columns I want to keep
```

```python
#finding the count of NaN values in each column
nan_counts = florida_df.isna().sum()

#Display the counts of NaN values for each column
print(nan_counts)
```

```python
#list of columns to keep (relevant to my story)
columns_to_keep = ['State_Name', 'County_Name', 'County_FIPS_Code', 'StudyYear', 'UNR_20to64', 'FUNR_20to64', 'MUNR_20to64', 'FLFPR_20to64_Under6',
                   'TotalPop', 'Households', 'H_Under6_BothWork', 'H_Under6_FWork', 'H_Under6_MWork', 'H_Under6_SingleM', 'MCInfant', 'MCInfant_flag', 
                   'MCToddler', 'MCToddler_flag', 'MCPreschool', 'MCPreschool_flag', '_75CInfant', '_75CInfant_flag', '_75CToddler', 
                   '_75CToddler_flag', '_75CPreschool', '_75CPreschool_flag']

#filter the DataFrame 
florida_df = florida_df[columns_to_keep]
```

```python
#printing the first 20 rows of the dataframe
florida_df.head(20)
```

```python
#finding the count of NaN values in each column
nan_counts2 = florida_df.isna().sum()

#Display the counts of NaN values for each column
print(nan_counts2)
```

### Observation: 
There are 76 rows out of 737 that contain all NaN values for all 12 of the last columns. These columns describe the childcare prices for each age group. This data is important to the story and these rows should evaluated on whether they should be dropped...

```python
# Filter the DataFrame to show all rows with NaN values
rows_with_nan = florida_df[florida_df.isna().any(axis=1)]

# Display the filtered DataFrame
rows_with_nan
```

### Observation: 
The 76 rows with 12 columns of NaN values are spread out throughout different counties in Florida. Therefore, not just one county's data will be affected if these rows are dropped. 

```python
#drop rows with any NaN values, in place
florida_df.dropna(inplace=True)

#display the modified DataFrame
florida_df
```
